```python
# Importar y organizar los datos
path = '/content/consumo-agua.csv'
df_consumo = pd.read_csv(path, header=0)
print(df_consumo["indice_des"][len(df_consumo)-1])
df_consumo.head(1)
```

# Preprocesamiento

```python
# cambiar indice de desarrollo por valor numerico
j = 0
for i in range(len(df_consumo)):
  if df_consumo["alcaldia"][i].find("GUSTAVO A. MADERO") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.832)
    j += 1
  elif df_consumo["alcaldia"][i].find("AZCAPOTZALCO") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.854)
    j += 1
  elif df_consumo["alcaldia"][i].find("BENITO JUAREZ") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.944)
    j += 1
  elif df_consumo["alcaldia"][i].find("MIGUEL HIDALGO") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.917)
    j += 1
  elif df_consumo["alcaldia"][i].find("CUAUHTEMOC") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.878)
    j += 1
  elif df_consumo["alcaldia"][i].find("VENUSTIANO CARRANZA") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.846)
    j += 1
  elif df_consumo["alcaldia"][i].find("CUAJIMALPA") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.850)
    j += 1
  elif df_consumo["alcaldia"][i].find("ALVARO OBREGON") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.836)
    j += 1
  elif df_consumo["alcaldia"][i].find("COYOACAN") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.883)
    j += 1
  elif df_consumo["alcaldia"][i].find("IZTAPALAPA") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.813)
    j += 1
  elif df_consumo["alcaldia"][i].find("MAGDALENA CONTRERAS") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.815)
    j += 1
  elif df_consumo["alcaldia"][i].find("TLALPAN") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.842)
    j += 1
  elif df_consumo["alcaldia"][i].find("XOCHIMILCO") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.809)
    j += 1
  elif df_consumo["alcaldia"][i].find("TLAHUAC") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.805)
    j += 1
  elif df_consumo["alcaldia"][i].find("MILPA ALTA") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.768)
    j += 1
  elif df_consumo["alcaldia"][i].find("IZTACALCO") != -1:
    df_consumo['indice_des'] = df_consumo['indice_des'].replace([df_consumo['indice_des'][i]],0.861)
    j += 1

print(df_consumo["indice_des"][len(df_consumo)-1])
print(len(df_consumo) - j)
```

```python
# separar df por grupos y poner los nuevos dfs en una lista
def split(df, group):
  gb = df.groupby(group)
  return [gb.get_group(x) for x in gb.groups]
s = split(df_consumo, "bimestre")
print(len(s))
```

```python
# generar nuevos dfs por bimestre
df_cons_b1 = s[0]
df_cons_b2 = s[1]
df_cons_b3 = s[2]
print(len(df_cons_b1))
print(len(df_cons_b2))
print(len(df_cons_b3))
```

```python
tomas.sort_values
alc_index = []
for i in range(len(tomas)):
  alc_index.append(tomas.index[-i-1])
#print(tomas[alc_index[0]])
ctotal_reg = []
tomas_reg = []
for i in range(len(alc_index)):
  ctotal_reg.append(consumo_total[alc_index[i]])
  tomas_reg.append(tomas[alc_index[i]])
print(alc_index)
print(ctotal_reg)
print(tomas_reg)
```

# Regresión Lineal

```python
import numpy as np
from sklearn.linear_model import LinearRegression

x = np.array(tomas_reg).reshape((-1, 1))
y = np.array(ctotal_reg)

model = LinearRegression()
model = LinearRegression().fit(x, y)

print('coefficient of determination:', model.score(x, y))
print('intercept:', model.intercept_)
print('slope:', model.coef_[0])

y_0 = model.intercept_
slope = model.coef_[0]

def f(x):
  return (x * slope) + y_0

print(f(0))

x_axis = list(range(3600))
line = list(map(f, x_axis))
```

```python
# Regresión lineal de las tomas vs consumo de agua
fig, ax = plt.subplots()
fig.canvas.draw()

# utilizar las alcaldías como ticks en el eje x
def update_ticks(x, pos):
  if x in tomas_reg:
    return alc_index[tomas_reg.index(x)]
  else:
    return x

ax.xaxis.set_ticks([100, 218, 321, 653, 803, 1064, 1352, 1663, 1725, 1989, 2307, 3258, 3521, 3600])
plt.xticks(rotation='vertical')
ax.xaxis.set_major_formatter(plt.FuncFormatter(update_ticks))


# puntos reales
plt.scatter(tomas_reg, ctotal_reg)
plt.plot(2307, 5885682.5, color='green', marker="s", markersize=12)
plt.plot(3521, 2693761.5, color='red', marker="s", markersize=12)

# predccion (modelo)
plt.plot(line, color="orange")

# proyeccion al eje x
plt.axvline(x = 2307, ymin = 0, ymax = .95,
            color ='green', linestyle='--', alpha = .5)
plt.axvline(x = 3521, ymin = 0, ymax = .82,
            color ='red', linestyle='--', alpha = .5)

plt.xlabel("Cantidad de tomas de agua por alcaldía")
plt.ylabel("Consumo de agua por alcaldía en millones de metros cúbicos")
plt.title('Consumo de agua por tomas de agua. Bimestre 1, CDMX 2019')
plt.tight_layout()
plt.savefig("linreg.pdf")
plt.show()
```
